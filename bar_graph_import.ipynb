```python
%pip install seaborn
%pip install xlrd

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.rc("font", family="Malgun Gothic")

plt.rc("axes", unicode_minus=False)
```

```python
df_2023 = pd.read_excel("2023_import_world.xlsx")
df_2022 = pd.read_excel("2022_import_world.xlsx")
df_2021 = pd.read_excel("2021_import_world.xlsx")
df_2020 = pd.read_excel("2020_import_world.xlsx")
```

```python
columns = ['Reporter', 'Quantity']

df_2023 = df_2023[columns]
df_2023['Quantity'] = df_2023['Quantity'] / 1000
df_2023 = df_2023.sort_values(by='Quantity', ascending=False)
df_2023 = df_2023.iloc[:5]

df_2022 = df_2022[columns]
df_2022['Quantity'] = df_2022['Quantity'] / 1000
df_2022 = df_2022.sort_values(by='Quantity', ascending=False)
df_2022 = df_2022.iloc[:5]

df_2021 = df_2021[columns]
df_2021['Quantity'] = df_2021['Quantity'] / 1000
df_2021 = df_2021.sort_values(by='Quantity', ascending=False)
df_2021 = df_2021.iloc[:5]

df_2020 = df_2020[columns]
df_2020['Quantity'] = df_2020['Quantity'] / 1000
df_2020 = df_2020.sort_values(by='Quantity', ascending=False)
df_2020 = df_2020.iloc[:5]
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(x="Quantity", y="Reporter", data=df_2023, color='green')
plt.xlabel("수입 양 (톤)")
plt.ylabel("국가")
plt.title("2023년 상위 5개 국가별 수입 양")

max_quantity = df_2023["Quantity"].max()
plt.xlim(0, max_quantity * 1.1)

for i, v in enumerate(df_2023["Quantity"]):
    ax.text(v, i, f"{v:,.1f}", va='center', ha='left', fontsize=10)

plt.show()
```
