```python
%pip install seaborn
%pip install xlrd

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.rc("font", family="Malgun Gothic")

plt.rc("axes", unicode_minus=False)
```

```python
df_2023 = pd.read_excel("2023_export_world.xlsx")

df_2023.head()
```

```python
df_2022 = pd.read_excel("2022_export_world.xlsx")

df_2022.head()
```

```python
df_2021 = pd.read_excel("2021_export_world.xlsx")

df_2021.head()
```

```python
df_2020 = pd.read_excel("2020_export_world.xlsx")

df_2020.head()
```

```python
columns = ['Reporter', 'Year', 'Quantity']
df_2023 = df_2023[columns]
df_2023['Quantity'] = df_2023['Quantity'] / 1000
df_2023 = df_2023.sort_values(by='Quantity', ascending=False)
df_2023 = df_2023.iloc[:10]
df_2023.head(10)
```

```python
columns = ['Reporter', 'Year', 'Quantity']
df_2022 = df_2022[columns]
df_2022['Quantity'] = df_2022['Quantity'] / 1000
df_2022 = df_2022.sort_values(by='Quantity', ascending=False)
df_2022 = df_2022.iloc[:10]
df_2022.head(10)
```

```python
columns = ['Reporter', 'Year', 'Quantity']
df_2021 = df_2021[columns]
df_2021['Quantity'] = df_2021['Quantity'] / 1000
df_2021 = df_2021.sort_values(by='Quantity', ascending=False)
df_2021 = df_2021.iloc[:10]
df_2021.head(10)
```

```python
columns = ['Reporter', 'Year', 'Quantity']
df_2020 = df_2020[columns]
df_2020['Quantity'] = df_2020['Quantity'] / 1000
df_2020 = df_2020.sort_values(by='Quantity', ascending=False)
df_2020 = df_2020.iloc[:10]
df_2020.head(10)
```

```python
df_merge = pd.concat([df_2023, df_2022, df_2021, df_2020], ignore_index=True)
df_merge.head(40)
```

```python
top6 = df_2023['Reporter'].head(6).tolist()

#    6  
df_top6 = df_merge[df_merge['Reporter'].isin(top6)]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_top6, x="Year", y="Quantity", hue="Reporter", marker="o")
plt.xlabel("")
plt.ylabel("  ()")
plt.title(" 6     ")
plt.legend(title="", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
